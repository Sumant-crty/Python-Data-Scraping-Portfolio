### Set up your API key
To use the Gemini API, you'll need an API key. If you don't already have one, create a key in Google AI Studio.

In Colab, add the key to the secrets manager under the "ðŸ”‘" icon in the left panel. Give it the name `GOOGLE_API_KEY`. Then pass the key to the SDK:

```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

### Initialize the Generative Model
Before you can make any API calls, you need to initialize the Generative Model. You can choose different models based on your needs. For general text generation, `gemini-pro` is a good choice. For more creative and chat-like interactions, `gemini-2.5-flash-preview-04-17` might be suitable.

```python
# Initialize the Gemini API
gemini_model = genai.GenerativeModel('gemini-flash-latest') # Using a model from the list of available models
```

### Make an API call
Now you can make API calls. For example, to generate a poem:

```python
# List available models
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

# Configure API key
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize the Gemini API with a working model
gemini_model = genai.GenerativeModel('gemini-flash-latest')

# Generate a poem
response = gemini_model.generate_content('Write a short poem about a rainy day.')
print(response.text)
```

```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

# Configure API key
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize the Gemini API with a working model
gemini_model = genai.GenerativeModel('gemini-flash-latest')

# Get user input for the poem task
user_poem_prompt = input("What kind of poem would you like to generate? ")

# Generate a poem based on user input
response = gemini_model.generate_content(f'Write a short poem about {user_poem_prompt}.')
print(response.text)
```
